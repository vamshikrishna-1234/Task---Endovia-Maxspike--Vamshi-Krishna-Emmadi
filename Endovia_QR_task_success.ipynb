```python
#indicators.py -- functions
import pandas as pd
import numpy as np

def ema(series,period):
    return series.ewm(span=period,adjust=False).mean()

def macd(df,fast=12,slow=26,signal=9):
    df['ema_fast']=ema(df['close'],fast)
    df['ema_slow']=ema(df['close'],slow)
    df['macd']=df['ema_fast'] - df['ema_slow']
    df['macd_signal']=ema(df['macd'],signal)
    return df

def rsi(series,period=14):
    delta=series.diff()
    gain=(delta.where(delta>0,0)).rolling(window=period).mean()
    loss=(-delta.where(delta<0,0)).rolling(window=period).mean()
    rs=gain / loss
    return 100 - (100 / (1 + rs))

```

```python
#signal - making a robust compisite signal

def composite_signal(row,weights,threshold=0.5):
    #row['macd_signal'],row['rsi_signal'],row['inhouse_signal'],etc.
    score=(
        weights['inhouse'] * row['inhouse_signal'] +
        weights['macd'] * row['macd_signal'] +
        weights['rsi'] * row['rsi_signal']

    )
    if score >= threshold:
        return 'BUY'
    elif score <= -threshold:
        return 'SELL'
    else:
        return 'HOLD'
```

```python

#funtion to get atm options form both dfs
def get_atm_option(options,spot_price,expiry,option_type):
    #Filter for expiry and type
    df=options[(options['expiry_date']==expiry) &
                 (options['option_type']==option_type)]
    #Find strike closest to spot
    df=df.copy()
    df['diff']=abs(df['strike_price'] - spot_price)
    return df.loc[df['diff'].idxmin()] if not df.empty else None
```

```python
#defining ema,macd,rsi here
import pandas as pd

def ema(series,period):
    return series.ewm(span=period,adjust=False).mean()

def macd(df,fast=12,slow=26,signal=9):
    df['ema_fast']=ema(df['close'],fast)
    df['ema_slow']=ema(df['close'],slow)
    df['macd']=df['ema_fast'] - df['ema_slow']
    df['macd_signal']=ema(df['macd'],signal)
    return df

def rsi(series,period=14):
    delta=series.diff()
    gain=delta.where(delta>0,0).rolling(window=period).mean()
    loss=-delta.where(delta<0,0).rolling(window=period).mean()
    rs=gain / loss
    rsi=100 - (100 / (1 + rs))
    return rsi
```

```python
def composite_signal(row,weights,threshold=0.5):
    score=(
        weights.get('inhouse',0) * row.get('inhouse_signal',0) +
        weights.get('macd',0) * row.get('macd_signal',0) +
        weights.get('rsi',0) * row.get('rsi_signal',0)
    )
    if score >= threshold:
        return 'BUY'
    elif score <= -threshold:
        return 'SELL'
    else:
        return 'HOLD'
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split

def train_spot_model(df,feature_cols,target_col):
    X=df[feature_cols]
    y=df[target_col]
    X_train,X_test,y_train,y_test=train_test_split(
        X,y,test_size=0.3,shuffle=False
    )
    model=XGBClassifier(use_label_encoder=False,eval_metric='logloss')
    model.fit(X_train,y_train)
    return model,X_train,X_test,y_train,y_test
#training ml model to predict future price
```

```python


#---------------------------CORE BACKTEST STRATEGY-------------------------------------
import pandas as pd
import numpy as np

def get_atm_option(options,spot_price,expiry,option_type):
    df=options[(options['expiry_date']==expiry) & (options['option_type']==option_type)]
    if df.empty:
        return None
    df=df.copy()
    df['diff']=abs(df['strike_price'] - spot_price)
    return df.loc[df['diff'].idxmin()]   ######repeated

def run_backtest(spot_df,options_df,initial_capital=200000,stop_loss=0.015,take_profit=0.03,force_exit_time="15:15:00"):
    trades=[]
    equity=[initial_capital]
    position=None
    capital=initial_capital
    for idx,row in spot_df.iterrows():
        signal=row.get('composite_signal','HOLD')
        current_time=row['datetime'].time()
        if position is None and signal in ['BUY','SELL']:
            expiry=row.get('closest_expiry',None)
            if pd.isna(expiry):
                continue
            spot_price=row['close']
            option_type='PE' if signal=='BUY' else 'CE'
            option=get_atm_option(options_df,spot_price,expiry,option_type)
            if option is not None:
                entry_price=option['close']
                position={
                    'entry_time': row['datetime'],
                    'strike': option['strike_price'],
                    'option_type': option_type,
                    'expiry': expiry,
                    'entry_price': entry_price,
                    'direction': signal,
                    'entry_index': idx
                }
        elif position is not None:
            current_option=get_atm_option(options_df,row['close'],position['expiry'],position['option_type'])
            if current_option is None:
                continue
            current_price=current_option['close']
            change=(current_price - position['entry_price']) / position['entry_price']
            if position['direction']=='SELL':
                change=-change
            exit_trade=False
            if change <= -stop_loss or change >= take_profit:
                exit_trade=True
            if current_time >= pd.to_datetime(force_exit_time).time():
                exit_trade=True

            if exit_trade:
                pnl=(position['entry_price'] - current_price) if position['direction']=='BUY' else (current_price - position['entry_price'])
                pnl *= 100  #Lot size,adjust for your symbol
                capital += pnl
                trades.append({
                    'entry_time': position['entry_time'],
                    'exit_time': row['datetime'],
                    'strike': position['strike'],
                    'option_type': position['option_type'],
                    'entry_price': position['entry_price'],
                    'exit_price': current_price,
                    'pnl': pnl
                })
                position=None
                equity.append(capital)
    equity_df=pd.DataFrame({'equity': equity})
    trades_df=pd.DataFrame(trades)
    metrics_df=compute_metrics(equity_df)
    return equity_df,trades_df,metrics_df

def compute_metrics(equity_df):
    returns=equity_df['equity'].pct_change().dropna()
    sharpe_ratio=returns.mean() / returns.std() * (252 ** 0.5)
    drawdown=(equity_df['equity'].cummax() - equity_df['equity']) / equity_df['equity'].cummax()
    max_drawdown=drawdown.max()
    total_return=(equity_df['equity'].iloc[-1] / equity_df['equity'].iloc[0]) - 1
    metrics={
        'sharpe_ratio': sharpe_ratio,
        'max_drawdown': max_drawdown,
        'total_return': total_return
    }
    return pd.DataFrame([metrics])
```

```python
import matplotlib.pyplot as plt

def plot_equity_curve(equity_series,out_file):
    plt.figure(figsize=(10,5))
    plt.plot(equity_series)
    plt.title("Equity curve")
    plt.xlabel("Time")
    plt.ylabel("Equity")
    plt.grid(True)
    plt.savefig(out_file)
    plt.close()

def plot_drawdown(equity_series,out_file):
    running_max=equity_series.cummax()
    drawdown=equity_series / running_max - 1
    plt.figure(figsize=(10,5))
    plt.plot(drawdown,color='red')
    plt.title("Drawdown")
    plt.xlabel("Time")
    plt.ylabel("Drawdown")
    plt.grid(True)
    plt.savefig(out_file)
    plt.close()
```

```python
from tqdm import tqdm

def run_backtest(spot_df,options_df,initial_capital=200000,stop_loss=0.015,take_profit=0.03,force_exit_time="15:15:00"):
    trades=[]
    equity=[initial_capital]
    position=None
    capital=initial_capital

    for idx,row in tqdm(spot_df.iterrows(),total=spot_df.shape[0],desc="Running backtest"):
        signal=row.get('composite_signal','HOLD')
        current_time=row['datetime'].time()
        #Entry logic
        if position is None and signal in ['BUY','SELL']:
            expiry=row.get('closest_expiry',None)
            if pd.isna(expiry):
                continue
            spot_price=row['close']
            option_type='PE' if signal=='BUY' else 'CE'
            option=get_atm_option(options_df,spot_price,expiry,option_type)
            if option is not None:
                entry_price=option['close']
                position={
                    'entry_time': row['datetime'],
                    'strike': option['strike_price'],
                    'option_type': option_type,
                    'expiry': expiry,
                    'entry_price': entry_price,
                    'direction': signal,
                    'entry_index': idx
                }
        #Exit logic
        elif position is not None:
            current_option=get_atm_option(options_df,row['close'],position['expiry'],position['option_type'])
            if current_option is None:
                continue
            current_price=current_option['close']
            change=(current_price - position['entry_price']) / position['entry_price']
            if position['direction']=='SELL':
                change=-change
            exit_trade=False
            if change <= -stop_loss or change >= take_profit:
                exit_trade=True
            if current_time >= pd.to_datetime(force_exit_time).time():
                exit_trade=True

            if exit_trade:
                pnl=(position['entry_price'] - current_price) if position['direction']=='BUY' else (current_price - position['entry_price'])
                pnl *= 100  #Lot size,adjust if needed
                capital += pnl
                trades.append({
                    'entry_time': position['entry_time'],
                    'exit_time': row['datetime'],
                    'strike': position['strike'],
                    'option_type': position['option_type'],
                    'entry_price': position['entry_price'],
                    'exit_price': current_price,
                    'pnl': pnl
                })
                position=None
                equity.append(capital)

    equity_df=pd.DataFrame({'equity': equity})
    trades_df=pd.DataFrame(trades)
    metrics_df=compute_metrics(equity_df)
    return equity_df,trades_df,metrics_df
```
